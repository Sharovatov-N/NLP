```python
tokens_data_stem = []

for i in range(df.shape[0]):
    row_sentence_s = ' '.join(df.tweet_stemmed[i])
    tokens_data_stem.append(row_sentence_s)
    
df['tweet_stemmed'] = tokens_data_stem
```

```python
tokens_data_lem = []

for i in range(df.shape[0]):
    row_sentence_l = ' '.join(df.tweet_lemmatized[i])
    tokens_data_lem.append(row_sentence_l)
    
df['tweet_lemmatized'] = tokens_data_lem
```

### 1. 
Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.

● Игнорируем слова, частота которых в документе строго превышает порог 0.9 с

● Ограничим количество слов, попадающий в мешок, с помощью max_features =

● Исключим стоп-слова с помощью stop_words='english'.

● Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с

#### Создаем мешок слов с помощью CountVectorizer для 'tweet_stemmed'.

```python
count_vectorizer = CountVectorizer(stop_words='english', 
                                   analyzer='word', 
                                   max_df=0.9, 
                                   max_features = 1000, 
                                   binary=False
                                  )

# Создаем the Bag-of-Words модель
   
bag_of_words_s = count_vectorizer.fit_transform(df['tweet_stemmed'])
                
# Отобразим Bag-of-Words модель как DataFrame

feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words_s.toarray(), columns = feature_names)
```

```python
count_vectorizer = CountVectorizer(stop_words='english', 
                                   analyzer='word', 
                                   max_df=0.9, 
                                   max_features = 3000, 
                                   binary=False
                                  )

# Создаем the Bag-of-Words модель
   
bag_of_words_l = count_vectorizer.fit_transform(df['tweet_lemmatized'])
                
# Отобразим Bag-of-Words модель как DataFrame

feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words_l.toarray(), columns = feature_names)
```

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
count_vect = CountVectorizer(stop_words='english', 
                             analyzer='word', 
                             max_df=1.0, 
                             max_features = 30000,  
                             token_pattern=r'\w{1,}'
                             )
count_vect.fit(trainDF['text'])

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(valid_y, predictions)
```

```python
tfidf_vect = TfidfVectorizer(stop_words='english', 
                                   analyzer='word', 
                                   max_df=0.9, 
                                   max_features = 10000, 
                                   binary=False
                                  )
tfidf_vect.fit(trainDF['text'])

xtrain_tfidf =  tfidf_vect.transform(train_x)
xvalid_tfidf =  tfidf_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_tfidf, train_y)
predictions = classifier.predict(xvalid_tfidf)
predictions
```

```python
accuracy_score(valid_y, predictions)
```

```python
vct = HashingVectorizer(analyzer='word', token_pattern=r'\w{1,}', n_features=1000)
vct.fit(trainDF['text'])

xtrain_hash =  vct.transform(train_x)
xvalid_hash =  vct.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_hash, train_y)
predictions = classifier.predict(xvalid_hash)
predictions
```

```python
accuracy_score(valid_y, predictions)
```
